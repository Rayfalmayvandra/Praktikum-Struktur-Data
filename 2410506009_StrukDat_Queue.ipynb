# **Implementasi Queue menggunakan List**

```python
class QueueList:
 def __init__(self):
    self.queue = []

 def enqueue(self, item):
    self.queue.append(item)

 def dequeue(self):
      if not self.is_empty():
          return self.queue.pop(0)
      return None

 def front(self):
    return self.queue[0] if not self.is_empty() else None

 def rear(self):
    return self.queue[-1] if not self.is_empty() else None

 def is_empty(self):
    return len(self.queue) == 0

 def size(self):
    return len(self.queue)

# Contoh penggunaan
queue = QueueList()
queue.enqueue(10)
queue.enqueue(20)
queue.enqueue(30)
print("Queue setelah enqueue:", queue.queue)
print("Dequeued element:", queue.dequeue())
print("Front element:", queue.front())
print("Rear element:", queue.rear())
print("Queue size:", queue.size())
```

# **Implementasi Queue menggunakan deque**

```python
from collections import deque

class QueueDeque:
  def __init__(self):
      self.queue = deque()

  def enqueue(self, item):
      self.queue.append(item)

  def dequeue(self):
      return self.queue.popleft() if not self.is_empty() else None

  def front(self):
      return self.queue[0] if not self.is_empty() else None

  def rear(self):
      return self.queue[-1] if not self.is_empty() else None

  def is_empty(self):
      return len(self.queue) == 0

  def size(self):
      return len(self.queue)

# Contoh penggunaan
queue = QueueDeque()
queue.enqueue(10)
queue.enqueue(20)
queue.enqueue(30)
print("Queue setelah enqueue:", list(queue.queue))
print("Dequeued element:", queue.dequeue())
print("Front element:", queue.front())
print("Rear element:", queue.rear())
print("Queue size:", queue.size())
```

# **Implementasi Queue dengan queue.Queue**

```python
from queue import Queue

queue = Queue()

queue.put(10)
queue.put(20)
queue.put(30)

print("Queue size:", queue.qsize())
print("Dequeued element:", queue.get())
print("Queue size setelah dequeue:", queue.qsize())
```

# **Implementasi Queue dengan Linked List**

```python
class Node:
  def __init__(self, data):
      self.data = data
      self.next = None

class QueueLinkedList:

  def __init__(self):
      self.front = self.rear = None

  def enqueue(self, data):
    new_node = Node(data)
    if self.rear is None:
        self.front = self.rear = new_node
    return
    self.rear.next = new_node
    self.rear = new_node

  def dequeue(self):
    if self.front is None:
      return None
    dequeued_element = self.front.data
    self.front = self.front.next
    if self.front is None:
      self.rear = None
      return dequeued_element

  def is_empty(self):
      return self.front is None

  def front_element(self):
      return self.front.data if self.front else None

  def rear_element(self):
      return self.rear.data if self.rear else None

queue = QueueLinkedList()
queue.enqueue(10)
queue.enqueue(20)
queue.enqueue(30)
print("Dequeued element:", queue.dequeue())
print("Front element:", queue.front_element())
print("Rear element:", queue.rear_element())
```

# **Latihan 1**

```python
class AntrianTiket:
  def __init__(self, capacity = 10):
    self.queue = []
    self.capacity = capacity

  def is_empty(self):
    return len(self.queue) == 0

  def size(self):
    return len(self.queue)

  def enqueue(self, ticket_id):
      if '7' in ticket_id:
        print(f"VIP Customer ticket {ticket_id} anda telah mendapatkan kursi!")
      else:

          if self.size() >= self.capacity:
            print(f"Antrian Penuh! {ticket_id} harap menunggu sebentar.")
          else:
              self.queue.append(ticket_id)
              print(f"Pelanggan {ticket_id} silahkan masuk ke antrian")

  def dequeue(self):
    if not self.is_empty():
      pelanggan = self.queue.pop(0)
      print(f"Pelanggan {pelanggan} telah membeli tiket.")
    else:
      print("Antrian kosong, tidak ada pelanggan yang bisa dilayyani.")

  def front(self):
    if not self.is_empty():
      print(f"Pelanggan pertama dalam antrian: {self.queue[0]}")
    else:
      print("Antrian kosong, tidak ada pelanggan.")

  def rear(self):
    if not self.is_empty():
      print(f"Pelanggan terakhir dalam antrian: {self.queue[-1]}")
    else:
      print("Antrian kosong, tidak ada pelanggan.")

  def show_queue(self):
      if self.is_empty():
        print("Antrian Kosong.")
      else:
        print("\n Antrian saat ini","->".join(self.queue))

antrian = AntrianTiket()

while True:
  print("\n<------------------------->")
  print("\nMENU TIKET:")
  print("1. Pesan Tiket")
  print("2. Layani Pelanggan")
  print("3. Lihat Antrian")
  print("4. Lihat Antrian Pertama")
  print("5. Lihat Antrian Terakhir")
  print("6. Keluar")
  print("\n<------------------------->")

  pilihan = input("Pilih Menu Tiket:")

  if pilihan == "1":
        ticket_id = input("Masukkan kode pelanggan: ").upper()
        antrian.enqueue(ticket_id)

  elif pilihan == "2":
        antrian.dequeue()

  elif pilihan == "3":
        antrian.show_queue()

  elif pilihan == "4":
        antrian.front()

  elif pilihan == "5":
        antrian.rear()

  elif pilihan == "6":
        print("Terima kasih! Program selesai.")
        break

  else:
        print("Pilihan tidak valid, silakan coba lagi.")
```

# **Latihan 2**

```python
class AntrianFotokopi:
  def __init__(self):
    self.queue = []

  def is_empty(self):
    return len(self.queue) == 0

  def size(self):
    return len(self.queue)

  def enqueue(self, order_id, num_pages):
      num_pages = int(num_pages)

      if num_pages < 10:
        print(f"Pesanan antrian {order_id} (dengan {num_pages} selesai, pesanan bisa langsung di ambil)")
        return

      if num_pages > 100:
          print(f"Pesanan {order_id} (dengan {num_pages} halaman) akan diproses dalam dua tahap.")
          self.queue.append((order_id, 100))
          self.queue.append((order_id, num_pages - 100))
      else:
          self.queue.append((order_id, num_pages))

      print(f"Pesanan {order_id} (dengan {num_pages} halaman) masuk ke antrian.")

  def dequeue(self):
    if not self.is_empty():
      pelanggan = self.queue.pop(0)
      print(f"Pesanan Pelanggan dengan pesanan {pelanggan[0]} ({pelanggan[1]} halaman) telah selesai.")
    else:
      print("Antrian kosong, tidak ada pelanggan yang bisa dilayyani.")

  def front(self):
    if not self.is_empty():
      print(f"Pelanggan pertama dalam antrian: {self.queue[0][0]} ({self.queue[0][1]} halaman)")
    else:
      print("Antrian kosong, tidak ada pelanggan.")

  def rear(self):
    if not self.is_empty():
      print(f"Pelanggan terakhir dalam antrian: {self.queue[-1][0]} ({self.queue[-1][1]} halaman)")
    else:
      print("Antrian kosong, tidak ada pelanggan.")

  def show_queue(self):
      if self.is_empty():
        print("Antrian Kosong.")
      else:
        print("\n Antrian saat ini","->".join(f"{order[0]} ({order[1]} halaman)" for order in self.queue))

antrian = AntrianFotokopi()

while True:
  print("\n<------------------------->")
  print("\nMENU:")
  print("1. Pesan fotokopi")
  print("2. Layani Pelanggan")
  print("3. Lihat Antrian")
  print("4. Lihat Antrian Pertama")
  print("5. Lihat Antrian Terakhir")
  print("6. Keluar")
  print("\n<------------------------->")

  pilihan = input("Pilih Menu pelayanan:")

  if pilihan == "1":
        order_id = input("Masukkan ID Pemesanan sebagai penanda antrian: ").upper()
        num_pages = input("Masukkan jumlah halaman yang ingin dipesan:")

        if not num_pages.isdigit():
          print("Jumlah halaman harus berupa angka")
          continue

        antrian.enqueue(order_id, num_pages)

  elif pilihan == "2":
        antrian.dequeue()

  elif pilihan == "3":
        antrian.show_queue()

  elif pilihan == "4":
        antrian.front()

  elif pilihan == "5":
        antrian.rear()

  elif pilihan == "6":
        print("Terima kasih! Sampai Jumpa Kembali.")
        break

  else:
        print("Pilihan tidak valid, silakan coba lagi.")
```

# **Latihan 3**

```python
class AntrianLayananParkir:
  def __init__(self, capacity=15):
    self.queue = []
    self.capacity = capacity

  def is_empty(self):
    return len(self.queue) == 0

  def size(self):
    return len(self.queue)

  def enqueue(self, vehicle_number, is_electric):

      if self.size() >= self.capacity:
            print(f"Parkiran penuh! {vehicle_number} tidak bisa masuk dan harus menunggu.")
            return

      if is_electric:
          self.queue.insert(0, vehicle_number)
          print(f" Kendaraan Listrik {vehicle_number} mendapatkan parkiran, silahkan masuk!")
      else:
          self.queue.append(vehicle_number)
          print(f"Kendaraan {vehicle_number} masuk ke parkiran.")

  def dequeue(self):
    if not self.is_empty():
      Kendaraan = self.queue.pop(0)
      print(f"Kendaraan {Kendaraan} telah keluar dari parkiran.")

  def front(self):
    if not self.is_empty():
      print(f"Kendaraan pertama dalam parkiran: {self.queue[0]}")
    else:
      print("Parkiran kosong, tidak ada kendaraan.")

  def rear(self):
    if not self.is_empty():
      print(f"Kendaraan terakhir dalam parkiran: {self.queue[-1]}")
    else:
      print("Parkiran kosong, tidak ada kendaraan.")

  def show_queue(self):
      if self.is_empty():
        print("Parkiran Kosong.")
      else:
        print("\n Kendaraan dalam parkiran:", " -> ".join(self.queue))

parkiran = AntrianLayananParkir()

while True:
  print("\n<------------------------->")
  print("\nMENU PARKIR MALL:")
  print("1. Masuk Parkiran/Ambil Tiket")
  print("2. Keluar Parkiran")
  print("3. Lihat Kendaraan di Parkiran")
  print("4. Lihat Kendaraan Pertama")
  print("5. Lihat Kendaraan Terakhir")
  print("6. Keluar Tidak Jadi Parkir")
  print("\n<------------------------->")

  pilihan = input("Pilih Menu Parkiran:")

  if pilihan == "1":
        vehicle_number = input("Masukkan Nomor Kendaraan: ").upper()
        is_electric = input("Apakah Kendaraan anda kendaraan listrik? (y/n):").strip() == "y"
        parkiran.enqueue(vehicle_number, is_electric)


  elif pilihan == "2":
        parkiran.dequeue()

  elif pilihan == "3":
        parkiran.show_queue()

  elif pilihan == "4":
        parkiran.front()

  elif pilihan == "5":
        parkiran.rear()

  elif pilihan == "6":
        print("Terima kasih! Sampai Jumpa Kembali.")
        break

  else:
        print("Pilihan tidak valid, silakan coba lagi.")
```

# **Latihan 4**

```python
class AntrianMahasiswa:
    def __init__(self, max_capacity=20):
        self.regular_queue = []
        self.achiever_queue = []
        self.capacity = max_capacity

    def is_empty(self):
        return len(self.regular_queue) == 0 and len(self.achiever_queue) == 0

    def size(self):
        return len(self.regular_queue) + len(self.achiever_queue)

    def enqueue(self, student_name, is_achiever, is_national_competitor):
        if is_national_competitor:
            print(f"Mahasiswa {student_name} langsung mengambil sertifikat tanpa antre!")
            return

        if self.size() >= self.capacity:
            if is_achiever:
                self.achiever_queue.append(student_name)
                print(f"Mahasiswa {student_name} masuk ke jalur **Berprestasi**.")
            else:
                self.regular_queue.append(student_name)
                print(f"Mahasiswa {student_name} masuk ke jalur **Reguler**.")
        else:
            self.regular_queue.append(student_name)
            print(f"Mahasiswa {student_name} masuk ke antrian.")

    def dequeue(self):
        if not self.is_empty():
            if self.achiever_queue:
                student = self.achiever_queue.pop(0)
                print(f"Mahasiswa **Berprestasi** {student} telah dilayani.")
            else:
                student = self.regular_queue.pop(0)
                print(f"Mahasiswa {student} telah dilayani.")
        else:
            print("Antrian kosong, tidak ada mahasiswa yang bisa dilayani.")

    def show_queue(self):
        if self.is_empty():
            print("Antrian kosong.")
        else:
            print("\n**Antrian saat ini:**")
            if self.achiever_queue:
                print("**Jalur Berprestasi:**", " -> ".join(self.achiever_queue))
            if self.regular_queue:
                print("**Jalur Reguler:**", " -> ".join(self.regular_queue))

antrian = AntrianMahasiswa()

while True:
    print("\n<------------------------->")
    print("MENU ANTRIAN MAHASISWA:")
    print("1. Masuk Antrian")
    print("2. Layani Mahasiswa")
    print("3. Lihat Antrian")
    print("4. Keluar Program")
    print("<------------------------->")

    pilihan = input("Pilih Menu Antrian: ")

    if pilihan == "1":
        student_name = input("Masukkan Nama Mahasiswa: ").title()
        is_achiever = input("Apakah mahasiswa memiliki prestasi akademik? (Y/N): ").strip().lower() == "y"
        is_national_competitor = input("Apakah mahasiswa pernah ikut kompetisi tingkat nasional? (Y/N): ").strip().lower() == "y"
        antrian.enqueue(student_name, is_achiever, is_national_competitor)

    elif pilihan == "2":
        antrian.dequeue()

    elif pilihan == "3":
        antrian.show_queue()

    elif pilihan == "4":
        print("Terima kasih! Program selesai.")
        break

    else:
        print("Pilihan tidak valid, silakan coba lagi.")
```
